```python
from math import sin, pi
import matplotlib.pyplot as plt
```

```python
# начальные условия
x0 = 0
y0 = 1
xn = 4 * pi
def f(x, y):
    return -sin(x)
# переменные для вычислений
H = 0.1
X = [x0]
Y = [y0]
x = x0
y = y0
# расчёты
while x < xn:
    k1 = f(x, y)
    k2 = f(x + H / 2, y + H * k1 / 2)
    k3 = f(x + H / 2, y + H * k2 / 2)
    k4 = f(x + H, y + H * k3)
    y = y + H * (k1 + 2 * k2 + 2 * k3 + k4) / 6
    x = x + H
    Y.append(y)
    X.append(x)
# Вывод графика
plt.plot(X, Y)
```
